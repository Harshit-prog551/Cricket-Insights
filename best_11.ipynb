```python
with open('C:\\Users\\harsh\\Videos\\best_11_cricket_files\\t20_json_files\\t20_wc_match_results.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head(11)
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard' : 'match_id'}, axis =1 , inplace = True)
df_match.head()
```

```python
df_match.to_csv('match_summery.csv', index = False)
```

```python
match_id_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1']+ ' Vs ' + row['team2']
    key2 = row['team2']+ ' Vs ' + row['team1']

    match_id_dict[key1] = row['match_id']
    match_id_dict[key2] = row['match_id']

match_id_dict
```

```python
with open('C:\\Users\\harsh\\Videos\\best_11_cricket_files\\t20_json_files\\t20_wc_batting_summary.json') as f:
    data = json.load(f)
all_rec = []
for rec in data:
    all_rec.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_rec)
df_batting.head(11)
```

```python
df_batting.shape
```

```python
df_batting['Out/Not_Out'] = df_batting.dismissal.apply(lambda x: 'out' if len(x)>0 else 'Not Out')
df_batting.head()
```

```python
df_batting.drop(columns = ['dismissal'], inplace = True)
df_batting.head(11)
```

```python
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('â€', '') )
df_batting['batsmanName'] = df_batting.batsmanName.apply(lambda x: x.replace('\xa0', '') )
df_batting.head(11)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_id_dict)
df_batting.head()
```

```python
df_batting.to_csv('batting_summery.csv', index = False)
```
